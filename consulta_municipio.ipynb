```python
# Importando as bibliotecas necessárias
import geopandas as gpd
from shapely.geometry import Point, Polygon
```

```python
# Lendo a camada shapefile com os polígonos dos municípios e a armazenando na variável mun
mun = gpd.read_file('/content/drive/MyDrive/BA_Municipios_2022.shp', encoding = 'utf-8')

mun.head() # visualizando o topo da tabela de atributos da variável mun
```

```python
# Criando uma seleção por município que ficará armazenada na variável selec_mun
selec_mun = mun[mun['NM_MUN'] == 'Correntina'] # Para alterar o município basta atualizar o nome

selec_mun # visualizando o topo da tabela de atributos da variável selec_mun
```

```python
# Carregando a lista de pontos e a armazenando na variável pts

pts = gpd.read_file('/content/drive/MyDrive/pts_decreto.shp')

pts.head() # Visualizando o topo da tabela de atributops da variável pts
```

```python
# Setando e reprojetando o sistema de coordenadas da variável pts para ser o mesmo da variável mun

pts = pts.set_crs(epsg = '31983') # Setando o crs para o original da camada pts (crs que ela foi gerada)

pts = pts.to_crs(epsg = '4674') # Reprojetando o crs da variável pts para o mesmo crs da variável mun
```

```python
# Filtrando ponto na variável pts através do índice do ponto (caso a lista só tenha um único ponto o índice sempre será zero)
selec_pts = pts.iloc[95]

selec_pts # Visualizando o conteúdo da variável selec_pts (resultado do filtro)
```

```python
# Consultando a relação espacial entre o polígono do município e o ponto. Neste caso verificando se o polígono contém o ponto

selec_mun['geometry'].contains(selec_pts['geometry']) # Se o ponto estiver contido o resultado será True, do contrário, False
```

```python
# Consultando a relação espacial entre o ponto e o polígono do município. Neste caso verificando se o ponto está dentro do polígono

selec_pts['geometry'].within(selec_mun['geometry']) # Se o ponto estiver dentro do polígono o resultado será True, do contrário, False
```
